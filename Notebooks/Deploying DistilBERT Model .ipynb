# Streamlit App: Sentiment Analysis Deployment

```python
# ==============================================================
# IMPORT REQUIRED LIBRARIES
#
# pandas      -> data handling
# streamlit   -> web application deployment
# LabelEncoder -> convert text labels into numeric labels
# joblib      -> load trained machine learning models
# ==============================================================

import pandas as pd
import streamlit as st
from sklearn.preprocessing import LabelEncoder
import joblib
```

```python
# ==============================================================
# REAL-TIME CUSTOMER FEEDBACK SENTIMENT ANALYSIS SYSTEM
#
# PURPOSE:
# Deploy the trained sentiment analysis model as an interactive
# web application using Streamlit.
#
# USERS CAN:
# 1. Upload customer review dataset
# 2. Run sentiment prediction automatically
# 3. View sentiment distribution
# 4. Generate business insights automatically
#
# BUSINESS VALUE:
# Allows organisations to monitor customer satisfaction in
# real time and support decision-making.
# ==============================================================

# ==============================================================
# STREAMLIT APPLICATION TITLE
#
# PURPOSE:
# Display application name at top of dashboard
#
# WHY:
# Gives users immediate understanding of system purpose
# ==============================================================

st.title("Real-Time Customer Feedback Sentiment Analysis System")

st.write(
    """
    Upload customer feedback data and automatically classify
    sentiment into Positive, Negative, or Neutral categories.
    """
)


# ==============================================================
# LOAD TRAINED MODEL
#
# PURPOSE:
# Load previously trained sentiment classification model
#
# WHY:
# Instead of retraining every time, deployment loads saved
# production-ready model.
#
# EXAMPLE MODELS:
# - Logistic Regression
# - Naive Bayes
# - Linear SVM
# - DistilBERT
# ==============================================================

# Replace filename with your saved model
# Example:
# model = joblib.load("trained_log_reg_model.pkl")

# Placeholder
# model = joblib.load("your_saved_model.pkl")


# ==============================================================
# FILE UPLOAD SECTION
#
# PURPOSE:
# Allow user to upload CSV file containing customer reviews
#
# WHY:
# Enables bulk sentiment prediction for multiple customer
# reviews simultaneously.
# ==============================================================

uploaded_file = st.file_uploader(
    "Upload CSV File",
    type=["csv"]
)


# ==============================================================
# PROCESS FILE AFTER UPLOAD
# ==============================================================

if uploaded_file is not None:

    # ==========================================================
    # LOAD DATASET
    #
    # PURPOSE:
    # Read uploaded CSV file into DataFrame
    #
    # WHY:
    # DataFrame allows structured manipulation of review data
    # ==========================================================

    df = pd.read_csv(uploaded_file)

    st.subheader("Uploaded Dataset Preview")
    st.dataframe(df.head())


    # ==========================================================
    # DISPLAY AVAILABLE COLUMNS
    #
    # PURPOSE:
    # Verify available columns in uploaded dataset
    #
    # WHY:
    # Helps identify correct review text column
    # ==========================================================

    st.subheader("Available Columns")

    st.write(df.columns.tolist())


    # ==========================================================
    # DEFINE TEXT COLUMN
    #
    # PURPOSE:
    # Select column containing customer review text
    #
    # WHY:
    # Model requires text input for prediction
    #
    # CHANGE THIS IF COLUMN NAME IS DIFFERENT
    # ==========================================================

    text_column_name = "review"


    # ==========================================================
    # CHECK COLUMN EXISTS
    #
    # PURPOSE:
    # Prevent runtime error if wrong column selected
    # ==============================================================

    if text_column_name not in df.columns:

        st.error(
            f"Column '{text_column_name}' not found. "
            f"Please update text_column_name."
        )


    else:

        # ======================================================
        # INITIALISE PREDICTION STORAGE
        #
        # PURPOSE:
        # Store prediction results
        # ======================================================

        predictions = []


        # ======================================================
        # SENTIMENT PREDICTION LOOP
        #
        # PURPOSE:
        # Run sentiment prediction for every customer review
        #
        # WHY:
        # Enables batch processing of uploaded customer reviews
        # ======================================================

        for index, row in df.iterrows():

            try:

                # Extract review text
                text = row[text_column_name]


                # ==================================================
                # RUN MODEL PREDICTION
                #
                # Replace placeholder with actual model prediction
                #
                # Example:
                # predicted_label = model.predict([text])[0]
                # ==================================================

                predicted_label = "POSITIVE"     # placeholder


            except Exception as e:

                print(
                    f"Error processing row {index}: {e}"
                )

                predicted_label = "ERROR"


            predictions.append(predicted_label)


        # ======================================================
        # SAVE PREDICTIONS INTO DATAFRAME
        #
        # PURPOSE:
        # Store predicted sentiment labels
        # ======================================================

        df["predicted_sentiment"] = predictions


        # ======================================================
        # LABEL ENCODING
        #
        # PURPOSE:
        # Convert text labels into numerical format
        #
        # WHY:
        # Numerical labels help downstream analytics and model
        # monitoring systems.
        # ======================================================

        label_encoder = LabelEncoder()

        df["predicted_encoded"] = label_encoder.fit_transform(
            df["predicted_sentiment"]
        )


        # ======================================================
        # DISPLAY PREDICTION RESULTS
        #
        # PURPOSE:
        # Show sentiment prediction output to user
        # ======================================================

        st.subheader("Prediction Results")

        st.dataframe(df)


        # ======================================================
        # SENTIMENT DISTRIBUTION VISUALISATION
        #
        # PURPOSE:
        # Count number of Positive, Negative and Neutral reviews
        #
        # WHY:
        # Helps business understand customer opinion trend
        # ==============================================================

        st.subheader("📊 Sentiment Distribution Overview")

        sentiment_counts = df["predicted_sentiment"].value_counts()

        st.bar_chart(sentiment_counts)


        # ======================================================
        # BUSINESS INSIGHT GENERATION
        #
        # PURPOSE:
        # Automatically interpret sentiment results
        #
        # WHY:
        # Raw predictions alone are not useful for management.
        # Business insight transforms analytics into decisions.
        # ==============================================================

        st.subheader("💡 Key Business Insights")

        most_common = sentiment_counts.idxmax()

        st.write(

            f"""
            The dominant sentiment is **{most_common}**.

            This represents the overall customer experience trend
            detected within the uploaded feedback dataset.
            """
        )


        # ======================================================
        # DECISION SUPPORT RECOMMENDATIONS
        #
        # PURPOSE:
        # Provide management recommendations automatically
        #
        # WHY:
        # Demonstrates practical business application of AI model
        # ==============================================================

        st.write(

            """
            This insight can help organisations to:

            • Monitor customer satisfaction levels

            • Detect service delivery issues early

            • Identify negative customer experiences before escalation

            • Improve product quality and customer service strategy

            • Support data-driven operational decision-making
            """
        )


        # ======================================================
        # OPTIONAL FILE EXPORT
        #
        # PURPOSE:
        # Save prediction results after analysis
        #
        # WHY:
        # Allows users to download analysed results
        # ==============================================================

        csv = df.to_csv(index=False).encode("utf-8")

        st.download_button(
            label="Download Prediction Results",
            data=csv,
            file_name="sentiment_analysis_results.csv",
            mime="text/csv"
        )


# ==============================================================
# PIPELINE SUMMARY
#
# DEPLOYMENT WORKFLOW:
#
# 1. Upload dataset
# 2. Detect review column
# 3. Run sentiment prediction model
# 4. Store prediction results
# 5. Encode sentiment labels
# 6. Visualise sentiment distribution
# 7. Generate business insights
# 8. Export analysed results
#
# BUSINESS OUTPUT:
# Real-time customer sentiment monitoring system
# ==============================================================
```
